```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from crewai import Agent, Task, Crew
```

```python
import os
# Set OpenAI key
os.environ["OPENAI_MODEL_NAME"] = "gpt-3.5-turbo"
```

```python
planner = Agent(
    role="Content Planner",
    goal="plan factual and accurte content on {topic}",
    backstory="You are working on planning  blog article about "
              "{topic}. You collect the informtion that will help audience "
              "specially women going through pregnancy learn something "
              "and can take necessary precutions and checkups periodically "
              "in that period. Plan a content in a such a way that it should "
              "cover what it is, its effect in pregnancy, cause, symptoms, "
              "diagnosis, its treatment. Also provide any externl links to read "
              "or follow. Also plan to add Maintaining emotional and physical "
              "during this period. Also include detail explanation of the abbrevted "
              "term with its splitting. Also, mention the specialized doctor(adjective) "
              "with whom they can consult",
    allow_delebrtion=False,
    verbose=True
)
```

```python
writer = Agent(
    role="Content Writer",
    goal="Write insightful and factually accurate "
         "piece about the topic: {topic}",
    backstory="You're working on a writing "
              "a new blog about the topic: {topic}. "
              "You base your writing on the work of "
              "the Content Planner and Introductor, who provides an outline "
              "and relevant context about the topic. "
              "You follow the main objectives and "
              "direction of the outline, "
              "as provide by the Content Planner. "
              "You also provide objective and impartial insights "
              "and back them up with information "
              "provide by the Content Planner. "
              "It should have sections as what it is, "
              "its effect in pregnancy, cause, symptoms, "
              "diagnosis, its treatment, externl links to read/follow "
              "Highlight the section which will have content on "
              "maintaining physical and emotional balence in this period "
              "Split if its an abbrevation and explain every bit of it."
              "Mention long form of it and define each abbrevted term."
              "Also add an introductory paragraph which is like a "
              "conversation to address audience about you are not alone "
              "in this surprising journey. And making them feel comfortable "
              "about the fact that they can get out of this really well. "
              "The content should be aesthetically comforting and catchy.",
    allow_delegation=False,
    verbose=True,
)
```

```python
editor = Agent(
    role="Editor",
    goal="Edit a given blog post to align with "
         "the writing style. ",
    backstory="You are an editor who receives a blog post "
              "from the Content Writer. "
              "Your goal is to review the blog post "
              "to ensure that it follows journalistic best practices,"
              "provides balanced viewpoints "
              "when providing opinions or assertions, "
              "and also avoids major controversial topics "
              "or opinions when possible."
              "The content should align with the sections given by "
              "Content Writer with its accurate contents",
    allow_delegation=False,
    verbose=True
)
```

```python
plan = Task(
    description=(
        "1. Identify the target audience, considering "
            "their situation and pain points.\n"
        "2. Develop a detailed content outline including "
            "what it is, its effect in pregnancy, cause, symptoms, "
            "diagnosis, its treatment, externl links to read/follow "
            "Highlight the section which will have content on "
            "maintaining physical and emotional balence in this period\n"
        "3. Include SEO keywords and relevant data or sources."
    ),
    expected_output="A comprehensive content plan document "
        "with an outline, given section analysis, "
        "SEO keywords, and resources.",
    agent=planner,
)
```

```python
write = Task(
    description=(
        "1. Use the content plan to craft a compelling "
            "blog post on {topic}.\n"
        "2. Incorporate SEO keywords naturally.\n"
		    "3. Sections/Subtitles are properly named "
            "in an engaging manner.\n"
        "4. Ensure the post is structured with an "
            "engaging introduction, insightful body, "
            "and a summarizing conclusion.\n"
        "5. Proofread for grammatical errors and\n"
    ),
    expected_output="A well-written blog post "
        "in markdown format, ready for publication, "
        "each section should have 1 or 2 paragraphs.",
    agent=writer,
)
```

```python
edit = Task(
    description=("Proofread the given blog post for "
                 "grammatical errors"),
    expected_output="A well-written blog post in markdown format, "
                    "ready for publication, "
                    "each section should have 1 or 2 paragraphs.",
    agent=editor
)
```

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=2
)
```

```python
result = crew.kickoff(inputs={"topic": "Everything you need to know about ITP in pregnancy"})
```

```python
from IPython.display import Markdown
Markdown(result)
```
